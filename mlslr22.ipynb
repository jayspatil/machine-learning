```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('../input/attendance/AttendanceMarksSA.csv')
```

```python
df.head()
```

```python
x = df['MSE']
y = df['ESE']
sns.scatterplot(x, y)
```

```python
beta0 = 0
beta1 = 1
alpha = 0.01
count = 10000
n = float(len(x))
```

```python
for i in range(count):
    Ybar = beta1*x + beta0
    beta1 = beta1 - (alpha/n)*sum(x*(Ybar-y))
    beta0 = beta0 - (alpha/n)*sum(Ybar-y)
    
print(beta0, beta1)
```

```python
endog = df['ESE']
exog = sm.add_constant(df[['MSE']])
print(exog)
```

```python
Ybar = beta1*x + beta0
plt.scatter(x,y)
plt.plot([min(x), max(x)], [min(Ybar), max(Ybar)], color= 'red') 
plt.show()
```

```python
import math
def RSE(y_true, y_predicted):
    y_true = np.array(y_true)
    y_predicted = np.array(y_predicted)
    RSS = np.sum(np.square(y_true - y_predicted))
    
    rse = math.sqrt(RSS/(len(y_true) -2))
    return rse

rse = RSE(df['ESE'], Ybar)
print(rse)
```
